```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/bykov-alexei/data-science-course/master/Pandas/perfomance.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.gender.value_counts().plot.pie();
```

```python
df.group.value_counts().plot.pie();
```

```python
df.groupby('test preparation course').agg({'math score': 'mean'}).plot.bar();
```

```python
df.groupby('group').agg({'writing score': 'mean'}).plot.bar();
```

```python
df.plot.scatter(x='math score', y='reading score');
```

```python
df.plot.scatter(x='math score', y='writing score');
```

```python
df.plot.scatter(x='reading score', y='writing score');
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x='math score', y='reading score', z='writing score', color='gender')
fig.show()
```
