```python
import pandas as pd
```

```python
url = "https://github.com/bykov-alexei/data-science-course/blob/master/Pandas/crime.csv?raw=true"
df = pd.read_csv(url, encoding='windows-1254')
```

```python
df.query("'2017-01-01' <= OCCURRED_ON_DATE < '2017-01-02' & OFFENSE_CODE_GROUP == 'Vandalism'").INCIDENT_NUMBER.size
```

```python
df.query("'2017-01-01' <= OCCURRED_ON_DATE < '2018-01-01' & OFFENSE_CODE_GROUP == 'Vandalism'")\
.groupby('OCCURRED_ON_DATE').count()
```
